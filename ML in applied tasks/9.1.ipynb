```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from KMeans_ext import KMeans_ex
```

```python
colors = ['r', 'y', 'b', 'g']
```

```python
import matplotlib.pyplot as plt

def true_scatter(df: pd.DataFrame) -> int:
    unique_classes = df['class'].unique()
    for ind, unique_class in enumerate(unique_classes):
        new_df = df[df['class'] == unique_class]
        x, y = new_df['x'], new_df['y']
        plt.scatter(x, y, c=colors[ind])
    plt.show()
    return len(unique_classes)
```

```python
def predicted_scatter(df: pd.DataFrame, predicted_classes: np.ndarray) -> None:
    new_df = df.copy()
    new_df['class'] = predicted_classes
    true_scatter(new_df)
```

```python
def draw_scatters(df: pd.DataFrame, initsialization: np.ndarray) -> None:
    num_cluster = true_scatter(df)
    # k_means = KMeans(n_clusters=num_cluster, random_state=42)
    # k_means.fit(df[['x', 'y']])
    k_means = KMeans_ex(K=num_cluster, init=initsialization)
    k_means.fit(np.array(df[['x', 'y']]))
    labels = k_means.predict(np.array(df[['x', 'y']]))
    # predicted_scatter(df, k_means.labels_)
    predicted_scatter(df, labels)
```

```python
df1 = pd.read_csv("1.csv")
df1
```

```python
df1.rename(columns={'a0': 'x', 'a1': 'y'}, inplace=True)
df1
```

```python
init = np.array([[-4, 4],
                 [4, -4]])
draw_scatters(df1, initsialization=init)
```
