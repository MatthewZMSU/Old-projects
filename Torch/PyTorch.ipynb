Попробуем вычисления на GPU.

```python
X = torch.rand([500, 20, 400])
# X
```

```python
torch.cuda.is_available()
```

```python
calc_device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
calc_device
```

```python
X_cuda = X.to(calc_device)
# X_cuda
```

```python
%time Y = (X - 10 * X ** 20) ** 5
```

```python
%time Y_cuda = (X_cuda - 10 * X_cuda ** 20) ** 5
```
